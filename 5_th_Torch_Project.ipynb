# Kerakli kutubxonalarni chaqirib olish

```python
import torch
import numpy as np
```

# Malumotlarni tensor korinishida yuklab olish

```python
x_soat=torch.tensor([[1.0],[2.0],[3.0]])
y_baho=torch.Tensor([[2.0],[4.0],[6.0]])
```

# Class yordamida Modelni tuzib olish

```python
# (1)--> Class yordamida Modelni tuzib olish
class Model(torch.nn.Module):
  def __init__(self):
      # Bu yerda torch.nn.Module bu yerda Super class Pytorch
      super().__init__()
      # torch.nn.Linear(kirish,chiqish) chiziqli model yordamida
      self.linear=torch.nn.Linear(1,1) #1 ta kirish va 1 ta chiqish
    # Method yordamida togri hisoblash funksiyasini kiritamiz(forward pass)
  def forward(self,x):
      y_pred=self.linear(x)
      return y_pred  
```

# Bizning Model

```python
# Bizning model    
model=Model()
print(model)  
```

# Loss va Optimizer larni tanlab olish

```python
# (2)--> Loss va Optimizer larni tanlab olish
criterion=torch.nn.MSELoss(reduction='sum')
optimizer=torch.optim.SGD(model.parameters(),lr=0.01)
```

# Training

```python
# (3.1) Training 
for epoch in range(500):  #Epochlar soni
  y_pred=model(x_soat)

  # Loss ||| xatolikni hisoblash va chop qilish
  loss=criterion(y_pred,y_baho)
  print(f'Epoch:{epoch}|,Loss:{loss.item()}')

  optimizer.zero_grad() #Har bir epoch uchun grad ni 0 ga tenglashtirib olsin
   
  #(3.2) back propagation ||| Teskari hisoblash
  loss.backward()

  # (3.3)-->Step ||| w ning qiymatini yangilash
  optimizer.step()

  # Bashorat uchun [test] qiymat ||| Ushbu qiymatlarimiz Tensor bolishi shart
soat_test=torch.Tensor([[4.0]])
print('Bashorat (training dan keyin), 4 soat oqisa:', model.forward(soat_test).data[0][0].item())
```
