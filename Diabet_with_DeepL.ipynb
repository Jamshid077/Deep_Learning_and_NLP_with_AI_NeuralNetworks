```python
import numpy as np 
import pandas as pd 
import torch
```

```python
xy_data=np.loadtxt('/content/diabetes_notColumns.csv',delimiter=',',dtype=np.float32)
xy_data
```

# X va Y data larga ajratib olish

```python
# birmartalik cod buni ikkinchimarta yuritsa xatolik qaytaradi
x_data=torch.from_numpy(xy_data[:700,0:-1])
y_data=torch.from_numpy(xy_data[:700,[-1]])
```

```python
print(f'{x_data},{y_data}')
```

# Class tordamida Model tuzib olish 3 etap

```python
# ///->1 
class Model(torch.nn.Module):
  def __init__(self):
    super(Model,self).__init__()
    # Ciziqli Modellar
    self.linear1=torch.nn.Linear(8,6) # kirish 8 va chiqish 6
    self.linear2=torch.nn.Linear(6,4) #kirish 6 va chiqish 4 
    self.linear3=torch.nn.Linear(4,1) # Kirish 4 va chiqish 1
    # Aktivatsiya funksiyasi (Sigmoid && ReLU)
    self.Sigmoid=torch.nn.Sigmoid()
    self.relu=torch.nn.ReLU()
  # method yordamida forward ni belgilash 
  def forward(self,x):
    natija1=self.Sigmoid(self.linear1(x))
    natija2=self.Sigmoid(self.linear2(natija1))
    y_pred=self.Sigmoid(self.linear3(natija2))
    return y_pred
# Bizning Model
model=Model()
```

# Loss va Optimizerlarni tanlash 

```python
# Loss va Otimizerlarni tanlash
criterion=torch.nn.BCELoss(reduction='mean')
optimizer=torch.optim.SGD(model.parameters(),lr=0.01)
```

# Training 

```python
for epoch in range(20000): # epochlar soni 15000
  y_pred=model(x_data) # Forward (togri xisoblash)
  loss=criterion(y_pred,y_data)
  if epoch%1000==0:  # loss ni hisoblash
    print(f'epoch # {epoch}--> Loss {loss.item():.4f}')
  optimizer.zero_grad() # Optimizer ni 0 ga tenglab olish 
  loss.backward() # Teskari hisoblash (Back Prop)
  optimizer.step() # Otimizer orqali w ni qiymatini yangilash 
print(f"\n Trainingdan song test qilib koramiz \n {'='*50}")
# T data
x_test = torch.from_numpy(xy_data[507:508, 0:-1])
print(f'test uchun data=={x_test}')
test=model(x_test)

print(f"\n Natija quyidagicha \n {'='*50}")
print(f'Bashorat qiymat: {test.item():.4f} | diabet mavjudligi:{test.item()>0.5}')

```
